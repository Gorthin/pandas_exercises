### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

### Step 4. How many products cost more than $10.00?

```python
chipo
#chipo.dtypes
```

```python
chipo['prices'] = chipo.item_price.apply(lambda x: float(str(x).replace("$","")))
#chipo.item_price = prices
chipo.loc[chipo['prices'] > 10, ['item_name', 'item_price']]
```

```python
# clean the item_price column and transform it in a float
prices = [float(value[1 : -1]) for value in chipo.item_price]

# reassign the column with the cleaned prices
chipo.item_price = prices

# delete the duplicates in item_name and quantity
chipo_filtered = chipo.drop_duplicates(['item_name','quantity','choice_description'])

# chipo_filtered

# select only the products with quantity equals to 1
chipo_one_prod = chipo_filtered[chipo_filtered.quantity == 1]
chipo_one_prod

# chipo_one_prod[chipo_one_prod['item_price']>10].item_name.nunique()
# chipo_one_prod[chipo_one_prod['item_price']>10]



#chipo.query('price_per_item > 10').item_name.nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
#chipo_filtered = chipo.drop_duplicates(['item_name','quantity'])
#chipo_filtered
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity']).loc[:, ['item_price', 'item_name']]
chipo_filtered.sort_values(by = 'item_name')
```

### Step 6. Sort by the name of the item

```python
chipo_sorted = chipo_filtered.item_name.sort_values()
chipo_sorted
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.groupby('item_name')
chipo.sum()
chipo.sort_values(by ='quantity', ascending = False).head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo.loc[chipo['item_name'] == 'Veggie Salad Bowl'].sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo.loc[(chipo['item_name'] == 'Canned Soda') & (chipo.quantity > 1)].count()
```
